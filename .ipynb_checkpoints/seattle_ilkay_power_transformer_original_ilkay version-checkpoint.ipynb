```python
from sklearn.cluster import KMeans
import pandas as pd


# Select the zip code column and drop missing values
zip_codes = df['zipcode'].dropna()

# Initialize a k-means clustering model with 10 clusters
kmeans = KMeans(n_clusters=7)

# Fit the model to the zip codes
kmeans.fit(zip_codes.values.reshape(-1, 1))

# Get the cluster labels for each zip code
cluster_labels = kmeans.labels_

# Add the cluster labels as a new column to the original dataset
df['zip_code_cluster'] = cluster_labels

# Save the preprocessed data to a new file
#df.to_csv('preprocessed_data.csv', index=False)
df
```

```python
#USING BOTH TRANSFORMERS AT THE SAME TIME FOR ANOTHER SHEET 

#The reason we use both the PowerTransformer and StandardScaler is to ensure that the data is both
#transformed to a normal distribution and standardized. The PowerTransformer is used to transform the data 
#to a normal distribution, while the StandardScaler is used to standardize the data, which is important for 
#some machine learning algorithms to perform well.

#Therefore, using both transformers ensures that the data is transformed to a normal distribution and 
#standardized at the same time, which can improve the performance of machine learning algorithms.

import pandas as pd
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PowerTransformer, StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score



# Define the columns to transform with PowerTransformer
columns_to_transform = X_train.columns

# Define the transformers
power_transformer = PowerTransformer()
scaler = StandardScaler()

# Define the column transformer
preprocessor = ColumnTransformer(
    transformers=[
        ('power', power_transformer, columns_to_transform),
        ('scale', scaler, columns_to_transform)
    ])

# Define the pipeline with the column transformer and the linear regression model
pipe = Pipeline([
    ('preprocessor', preprocessor),
    ('regressor', LinearRegression())
])

# Fit the pipeline on the training data
pipe.fit(X_train, y_train)

# Make predictions on the test data
y_pred = pipe.predict(X_test)

# Compute the R2 score
r2 = r2_score(y_test, y_pred)
print('R2 score:', r2)
```

```python
#only power transformer 

# Import necessary libraries
import pandas as pd
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PowerTransformer
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split


# Define the transformer for all columns
transformer = ColumnTransformer(
    transformers=[
        ('num', PowerTransformer(), X.columns)
    ])

# Define the pipeline with transformer and linear regression model
pipe = Pipeline([
    ('transformer', transformer),
    ('regressor', LinearRegression())
])

# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Fit the pipeline on the training data
pipe.fit(X_train, y_train)

# Predict on the test data
y_pred = pipe.predict(X_test)

# Evaluate the model
score = pipe.score(X_test, y_test)
print(f"R-squared score: {score:.3f}")
```
