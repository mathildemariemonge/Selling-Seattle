```python
CORR_THRESH = 0.80
corr_matrix=df.corr().abs()  #abs stands for absolute
upper_triangle=corr_matrix.where(np.triu(np.ones(corr_matrix.shape),k=1).astype(bool))
corrd_cols = [column for column in upper_triangle.columns if any(upper_triangle[column] > CORR_THRESH)]
df.drop(corrd_cols,axis=1,inplace=True)
df.columns
```

```python
from sklearn.cluster import KMeans
import pandas as pd


# Select the zip code column and drop missing values
zip_codes = df['zipcode'].dropna()

# Initialize a k-means clustering model with 10 clusters
kmeans = KMeans(n_clusters=7)

# Fit the model to the zip codes
kmeans.fit(zip_codes.values.reshape(-1, 1))

# Get the cluster labels for each zip code
cluster_labels = kmeans.labels_

# Add the cluster labels as a new column to the original dataset
df['zip_code_cluster'] = cluster_labels

# Save the preprocessed data to a new file
#df.to_csv('preprocessed_data.csv', index=False)
df
```

```python
zip_dummed = pd.get_dummies(df['zip_code_cluster'], prefix='zip_code_cluster_dummy')
zip_dummed
```

```python
df = pd.concat([df, zip_dummed], axis=1)
```

```python
df.drop('zip_code_cluster', axis=1, inplace=True)
df.drop(['zipcode'], axis =1, inplace = True)
df
```

```python
from sklearn.preprocessing import PowerTransformer
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# Assuming X and y are already preprocessed and split into training and test sets

# Create PowerTransformer object
pt = PowerTransformer()

# Fit and transform X_train with PowerTransformer
X_train_pt = pt.fit_transform(X_train)

# Transform X_test using the fitted PowerTransformer object
X_test_pt = pt.transform(X_test)


```

```python
# Create linear regression object and fit on the training data
lin_reg = LinearRegression()
lin_reg.fit(X_train_pt, y_train)

# Predict on the test data
y_pred = lin_reg.predict(X_test_pt)
```

```python
# Calculate RMSE
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print("RMSE: ", rmse)
```
